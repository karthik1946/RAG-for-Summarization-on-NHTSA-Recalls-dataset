```python
import pandas as pd
raw = pd.read_csv('/kaggle/input/recall/FLAT_RCL 3.txt', sep='\t', encoding='latin1', names=list(range(27)), low_memory=False, on_bad_lines='skip')
```

```python
manufacturers = ['BMW','MINI','MERCEDES','LEXUS','NISSAN','AUDI','VOLKSWAGEN','HYUNDAI','INFINITI','KIA','MAZDA',
                'LAND ROVER','VOLVO','HONDA','JAGUAR','ACURA','TOYOTA','MOPAR','SUBARU','PORSCHE','FORD']

names = ['RECORD_ID','CAMPNO','MAKETXT','MODELTXT','YEARTXT','MFGCAMPNO','COMPNAME','MFGNAME','BGMAN','ENDMAN','RCLTYPECD',
        'POTAFF','ODATE','INFLUENCED_BY','MFGTXT','RCDATE','DATEA','RPNO','FMVSS','DESC_DEFECT','CONEQUENCE_DEFECT',
        'CORRECTIVE_ACTION','NOTES','RCL_CMPT_ID','MFR_COMP_NAME','MFR_COMP_DESC','MFR_COMP_PTNO']

# keep only relevant car manufacturers
data = raw[raw[2].isin(manufacturers)]

#rename columms 
data.columns = names

# drop 1 record in POTAFF column
data.dropna(subset=['POTAFF'], axis=0, inplace=True)

# inspect the data
print('Relevant data dimensions: ', data.shape, '\n')
print(data.info())
data.head(n=3)
```

```python
data = data[data['MAKETXT'].isin(['FORD', 'TOYOTA'])]
```

```python
relevant_columns = ['MAKETXT', 'MODELTXT', 'YEARTXT', 'DESC_DEFECT', 'CONEQUENCE_DEFECT', 'CORRECTIVE_ACTION']
data = data[relevant_columns]
```

```python
data.columns = ['make', 'model', 'year', 'defect_summary', 'consequence_summary', 'corrective_summary']
```

```python
data['combined_summary'] = data['defect_summary'] + " " + data['consequence_summary'] + " " + data['corrective_summary']
```

```python
data['combined_summary'] = data['combined_summary'].fillna('missing information')
data['combined_summary'] = data['combined_summary'].astype(str)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

input_query = {
    'make': 'ford',
    'model': 'escape',
    'year': '2001',
    'issue': 'stuck throttle risk'
}

query_text = f"{input_query['make']} {input_query['model']} {input_query['year']} {input_query['issue']}"

query_embedding = embedding_model.encode([query_text])
similarities = cosine_similarity(query_embedding, embeddings)
most_similar_idx = similarities.argmax()
relevant_record = data.iloc[most_similar_idx]
```

```python
data['combined_text'] = data['combined_text'].fillna("").astype(str)
embedding_model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
embeddings = embedding_model.encode(data['combined_text'].tolist(), show_progress_bar=True)
query_text = (
    f"MAKE: [{input_query['make']}] | " +
    f"MODEL: [{input_query['model']}] | " +
    f"YEAR: [{input_query['year']}] | " +
    f"{input_query['issue']}"
)

query_embedding = embedding_model.encode([query_text])

weights = [2.0 if "MAKE" in text else 1.0 for text in data['combined_text']]
similarities = cosine_similarity(query_embedding, embeddings) * weights
```

```python
most_similar_idx = similarities.argmax()
relevant_record = data.iloc[most_similar_idx]

print("Relevant Record:", relevant_record)
```
